```python
##DATA CLEANING
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
print (df.columns)
```

```python
# Filling missing values in Age with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Filling missing values in Embarked with the most common value
df['Embarked_Q'].fillna(df['Embarked_Q'].mode()[0], inplace=True)
df['Embarked_S'].fillna(df['Embarked_S'].mode()[0], inplace=True)

# Filling missing values in Fare with the median
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# Dropping the Cabin column due to many missing values
df.drop(columns='Cabin', inplace=True)
```

```python
# Extract titles from names
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

df.drop(columns=['Name', 'Ticket'], inplace=True)
```

```python

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
# Encode categorical variables
df = pd.get_dummies(df, columns=['Sex_male', 'Embarked_Q', 'Embarked_S', 'Title'], drop_first=True)
df = pd.get_dummies(df, columns=['Pclass'], drop_first=True)
```

```python

df.drop(columns=['PassengerId'], inplace=True)
```

```python
# Saving the cleaned dataset
df.to_csv('cleaned_titanic.csv', index=False)
```
